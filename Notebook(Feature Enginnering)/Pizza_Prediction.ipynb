```python
X = df.drop(['order_details_id','order_id','pizza_id','order_time','order_date','total_price','unit_price'],axis=1)
y = df['total_price']
```

```python
lreg = LinearRegression()
sfs1 = sfs(lreg, k_features=5, forward=True, verbose=2, scoring='neg_mean_squared_error')
```

```python
cat_cols = X.select_dtypes('object')

for i in cat_cols:
  le  = LabelEncoder()
  X[i] = le.fit_transform(X[i])
  dump(le, i+'_joblib')
```

```python
sfs1.fit(X,y)
```

```python
feat_names = list(sfs1.k_feature_names_)
print('Forward Selection',feat_names)
```

```python
X
```

```python
X = X[feat_names]
X.head(5)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
lreg.fit(X,y)
```

```python
y_pred = lreg.predict(X_test)
```

```python
lreg.score(X_train,y_train),lreg.score(X_test,y_test)
```

```python
with open('Model.pkl','wb') as f:
  pickle.dump(lreg,f)
```
